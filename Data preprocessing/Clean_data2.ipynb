```python
fixture_data_path = r'..\Data\RAW\FBREF_Dataset\bundesliga_2022-2023_fixture_data.csv'
player_data_path = r'..\Data\RAW\FBREF_Dataset\bundesliga_2022-2023_player_data.csv'
```

```python
fixture_raw = pd.read_csv(fixture_data_path)
```

```python
fixture_raw.head(10)
```

```python
player_raw = pd.read_csv(player_data_path)
player_raw.head(10)
```

```python
columns_to_remove = [col for col in player_raw.columns if col.endswith('_#') or col.endswith('_Pos')]
count = 0
columns_to_keep = []
for col in columns_to_remove:
    if count >= 2:
        player_raw = player_raw.drop(col, axis=1)
    count += 1
```

```python
player_raw = player_raw.rename(columns={"Summary_#": "#", "Summary_Pos": "Pos"})
```

```python
player_raw.info(verbose=True)
```

```python
merged_player_raw = player_raw.merge(fixture_raw[['Home', 'Away', 'game_id', 'Wk', 'season', 'Date']], left_on='game_id', right_on='game_id', how='left')
merged_player_raw['Team Name'] = merged_player_raw.apply(lambda row: row['Home'] if row['home'] == 1 else row['Away'], axis=1)
merged_player_raw.drop(['Home', 'Away', 'index'], axis=1, inplace=True)

print(merged_player_raw['Age'])
merged_player_raw.head(15)
```

```python
clean_data_GK_path = r'..\Data\Clean\GK.csv'
clean_data_Team_path = r'..\Data\Clean\Team.csv'
clean_data_Player_path = r'..\Data\Clean\Player.csv'
```

```python
grouped_position = merged_player_raw.groupby(merged_player_raw.Pos)
df_GK = grouped_position.get_group("GK")

df_Team = merged_player_raw[merged_player_raw['Pos'].isna()]

df_Player = merged_player_raw[(merged_player_raw['Pos'] != "GK") & (merged_player_raw['Pos'].notna())]
```

```python
if not os.path.exists(clean_data_GK_path):
    df_GK.to_csv(clean_data_GK_path, index=False)
else:
    df_GK.to_csv(clean_data_GK_path, mode='a', header=False, index=False)



if not os.path.exists(clean_data_Team_path):
    df_Team.to_csv(clean_data_Team_path, index=False)
else:
    df_Team.to_csv(clean_data_Team_path, mode='a', header=False, index=False)


if not os.path.exists(clean_data_Player_path):
    df_Player.to_csv(clean_data_Player_path, index=False)
else:
    df_Player.to_csv(clean_data_Player_path, mode='a', header=False, index=False)
```

```python
checkplayer = checkplayer.assign(Pos=checkplayer['Pos'].str.split(',')).explode('Pos')

position_counts = checkplayer.groupby(['Player', 'Pos']).size().reset_index(name='Count')

# Find the most played position for each player
most_played_positions = position_counts.loc[position_counts.groupby('Player')['Count'].idxmax()]
# Merge back the most played position
checkplayer = checkplayer.merge(most_played_positions[['Player', 'Pos']], 
                                on=['Player'], 
                                how='left', 
                                suffixes=('', '_MostPlayed'))

# Replace positions
checkplayer['Pos'] = checkplayer['Pos_MostPlayed'].fillna(checkplayer['Pos'])

# Drop temporary columns
checkplayer.drop(columns=['Pos_MostPlayed'], inplace=True)
```

```python
# Filter the dataset for players who played in or after 2020
players = players[players['last_season'] >= 2020]

# Define the list of desired competition IDs
desired_competitions = ['ES1', 'FR1', 'IT1', 'GB1', 'L1']

# Filter the DataFrame
players = players[players['current_club_domestic_competition_id'].isin(desired_competitions)]
```
