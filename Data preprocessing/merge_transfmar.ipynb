```python
transfer = pd.read_csv('../Data/RAW/Transfermarkt_Dataset/transfers.csv')
transfer.head(10)
```

```python
transfer.info()
```

```python
# Filter the dataset for transfer after 2020
#transfer = transfer[(transfer['transfer_season'] >= 20/21)]
transfer['transfer_date'] = pd.to_datetime(transfer['transfer_date'])

# Filter the data within the date range
transfer = transfer[
    #(transfer['transfer_date'] >= '2017-01-01') & 
    (transfer['transfer_date'] <= '2025-12-31')
]
```

```python
transfer.info()
```

```python
transfer.isnull().sum()
```

```python
transfer.head()
```

```python
merged_df = pd.merge(players, transfer, 
                     left_on=['player_id'], 
                     right_on=['player_id'], 
                     how='left')
```

```python
merged_df.info()
```

```python
merged_df.isnull().sum()
```

```python
merged_df.head(10)
```

```python
dup_rows = merged_df[merged_df['player_id'].duplicated()]
dup_rows.head(10)
```

```python
nan_rows = merged_df[merged_df['player_name'].isna()]
```

```python
nan_rows.head()
```

```python
# Reformat the season
merged_df['transfer_season'] = merged_df['transfer_season'].astype(str).apply(lambda x: f"20{x[:2]}-20{x[3:]}")
```

```python
merged_df = merged_df.rename(columns={'to_club_name': 'club'})
```

```python
merged_df.drop(columns=['first_name', 'last_name', 'player_code', 'sub_position', 'position', 'url', 'agent_name', 'current_club_domestic_competition_id', 'from_club_id', 'from_club_name'], inplace=True)
```

```python
print(merged_df.loc[(merged_df['name'] == 'Aaron Wan-Bissaka')])
```

```python
# Create a mapping from club_id to current_club_name
club_id_to_name = merged_df.set_index('current_club_id')['current_club_name'].to_dict()

# Apply the mapping to rename the 'club' column
merged_df['club'] = merged_df['to_club_id'].map(club_id_to_name).fillna(merged_df['club'])

merged_df['club'].fillna(merged_df['current_club_name'], inplace=True)
# Filter rows where club is not NaN
non_nan_club = merged_df[merged_df['club'].notna()]

# Duplicate and modify the 'club' value
duplicated_rows = non_nan_club.copy()
duplicated_rows['club'] = duplicated_rows['current_club_name']

# Concatenate back to original DataFrame
merged_df = pd.concat([merged_df, duplicated_rows], ignore_index=True)
```

```python
merged_df.head()
```

```python
# Replace 'col1' and 'col2' with your chosen columns
key_columns = ['name', 'club']

# Check uniqueness in df1
is_unique_df1 = not merged_df.duplicated(subset=key_columns).any()


print(f"Unique in df1: {is_unique_df1}")
```

```python
duplicates_df2 = merged_df[merged_df.duplicated(subset=key_columns, keep=False)]
 
print("Duplicates in df1:")
print(duplicates_df2.count())
```
