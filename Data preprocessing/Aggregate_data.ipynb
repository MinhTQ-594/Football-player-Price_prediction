```python
clean_data_GK_path = '../Data/Clean/GK.csv'
clean_data_Team_path = '../Data/Clean/Team.csv'
clean_data_Player_path = '../Data/Clean/Player.csv'
```

```python
player_new = pd.read_csv(clean_data_Player_path)

# player_new['Age'] = player_new['Age'].convert_dtypes()

# print(player_new['Age'].dtypes)

# def convert_age(age_str):
#     years, days = map(int, age_str.split('-'))
#     return years + days / 365  # Assuming 365 days in a year for conversion

# # Apply the conversion to the 'age' column
# player_new['Age'] = player_new['Age'].apply(convert_age)

player_new.drop(['Age'], axis=1)
```

```python
player_new_group =  player_new.groupby(['Player', '#', 'Team Name', 'Nation', 'Pos', 'season']).agg({
    'Min': 'mean',
    'Gls': 'mean',
    'Ast': 'mean',
    'PK': 'sum',
    'PKatt': 'sum',
    'Sh': 'mean',
    'SoT': 'mean',
    'CrdY': 'sum',
    'CrdR': 'sum',
    'Touches': 'mean',
    'Tkl': 'mean',
    'Int': 'mean',
    'Blocks': 'mean',
    'xG': 'mean',
    'npxG': 'mean',
    'xAG': 'mean',
    'SCA': 'mean',
    'GCA': 'mean',
    'Cmp_x': 'mean',
    'Att_x': 'mean',
    'Cmp%_x': 'mean',
    'PrgP': 'mean',
    'Carries': 'mean',
    'PrgC': 'mean',
    'Att_x': 'mean',
    'Succ': 'mean',
    'game_id': 'count'       # Total minutes played
}).reset_index()

player_new_group = player_new_group.rename(columns={'game_id': 'total_games'})
```
