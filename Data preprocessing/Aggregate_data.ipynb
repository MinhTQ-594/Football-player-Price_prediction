```python
import pandas as pd
```

```python
player_new = pd.read_csv('../Data/Clean/Player.csv')

player_new['Age'] = player_new['Age'].convert_dtypes()

print(player_new['Age'].dtypes)

def convert_age(age_str):
    years, days = map(int, age_str.split('-'))
    return years + days / 365  # Assuming 365 days in a year for conversion

# Apply the conversion to the 'age' column
player_new['Age'] = player_new['Age'].apply(convert_age)
```

```python
player_new_group =  player_new.groupby(['Player', '#', 'Team Name', 'Nation', 'Pos', 'season']).agg({
    'Age' : 'mean',
    'Min': 'sum',
    'Gls': 'sum',
    'Ast': 'sum',
    'PK': 'sum',
    'PKatt': 'sum',
    'Sh': 'sum',
    'SoT': 'sum',
    'CrdY': 'sum',
    'CrdR': 'sum',
    'Touches': 'sum',
    'Tkl': 'sum',
    'Int': 'sum',
    'Blocks': 'sum',
    'xG': 'sum',
    'npxG': 'sum',
    'xAG': 'sum',
    'SCA': 'sum',
    'GCA': 'sum',
    'Cmp_x': 'sum',
    'Att_x': 'sum',
    'Cmp%_x': 'mean',
    'PrgP': 'sum',
    'Carries': 'sum',
    'PrgC': 'sum',
    'Att_x': 'sum',
    'Succ': 'sum',
    'home': 'mean',               # Proportion of home games
    'game_id': 'count'       # Total minutes played
}).reset_index()

player_new_group = player_new_group.rename(columns={'game_id': 'total'})
player_new_group = player_new_group.rename(columns={'home': 'portion_home_game'})
```
