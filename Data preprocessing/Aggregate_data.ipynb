```python
def find_duplicate_entries(df):
  """
  Finds rows in a DataFrame that have the same 'name' and 'team' but different '#' values.

  Args:
    df: The pandas DataFrame to analyze.

  Returns:
    A DataFrame containing the duplicate entries.
  """

  # Group by 'name' and 'team' and count the occurrences
  grouped_df = df.groupby(['Player', 'Team Name','season']).size().reset_index(name='count')

  # Filter for groups with more than one occurrence
  duplicate_groups = grouped_df[grouped_df['count'] > 1]

  # Merge with the original DataFrame to get the actual rows
  duplicate_rows = pd.merge(duplicate_groups, df, on=['Player', 'Team Name', 'season'], how='left')

  # Filter for rows with different '#' values within the same group
  duplicate_rows = duplicate_rows.groupby(['Player', 'Team Name', 'season'])['#'].nunique().gt(1).reset_index(name='has_different_hashes')
  duplicate_rows = duplicate_rows[duplicate_rows['has_different_hashes']]

  # Merge back with the original DataFrame to get the final result
  duplicate_entries = pd.merge(duplicate_rows, df, on=['Player', 'Team Name', 'season'], how='left')

  return duplicate_entries

# Example usage:
# Assuming 'df' is your DataFrame
duplicate_entries = find_duplicate_entries(player_new)
```

```python
import functools
def wavg(val, df, weight):
    try:
        w = df.loc[val.index, weight]
        return (val * w).sum() / w.sum()
    except ZeroDivisionError:
        return val.mean()
```
