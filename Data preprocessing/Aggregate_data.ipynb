```python
# Step 1: Create a DataFrame with unique keys
weight_df = player_new[['Wk', 'season']].drop_duplicates()

weight_df = weight_df.sort_values(by=['season', 'Wk'], ascending=[True, True])

# Step 2: Apply EWA to create weights
weight_df['weight'] = range(1, len(weight_df) + 1)
```

```python
# Merge back to original DataFrame
merged_df = player_new.merge(weight_df, on=['Wk', 'season'], how='left')
```

```python
import functools
def wavg(val, df, weight):
    try:
        w = df.loc[val.index, weight]
        return (val * w).sum() / w.sum()
    except ZeroDivisionError:
        return val.mean()
    
# Binding df to function with functools
fwavg = functools.partial(wavg, df=merged_df, weight='weight')
```
