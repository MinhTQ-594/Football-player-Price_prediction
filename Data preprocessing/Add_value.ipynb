```python
import pandas as pd
import os
from datetime import datetime 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
player = pd.read_csv('../Data/Clean/Player_Gr.csv')
player.head()
```

```python
players = pd.read_csv('../Data/RAW/Transfermarkt_Dataset/shortenplayers.csv')
```

```python
players.head()
```

```python
# Replace 'col1' and 'col2' with your chosen columns
key_columns = ['player_id']

# Check uniqueness in df1
is_unique_df1 = not players.duplicated(subset=key_columns).any()


print(f"Unique in df1: {is_unique_df1}")

is_unique_df2 = not player.duplicated(subset=key_columns).any()


print(f"Unique in df2: {is_unique_df2}")

```

```python
# Merge using the refined match
merged_df = pd.merge(players, player, 
                     left_on = key_columns, 
                     right_on = key_columns, 
                     how='right')
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df.isna().sum()
```

```python
merged_df.drop(columns=['first_name', 'last_name', 'player_code', 'sub_position', 'position', 'url', 'agent_name', 'last_season', 'current_club_id', 'country_of_citizenship'], inplace=True)
```

```python
merged_df.to_csv('../Data/Clean/Player_merge.csv', index=False)
```
